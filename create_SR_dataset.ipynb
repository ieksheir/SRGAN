```python
# Define the custom loss function
class VGGLoss(tf.keras.Model):
    def __init__(self, feature_layer=36):
        super(VGGLoss, self).__init__()
        self.vgg = tf.keras.applications.VGG19(include_top=False, input_shape=(None, None, 3))
        self.vgg.trainable = False
        self.feature_layer = feature_layer
        self.loss = tf.keras.losses.MeanSquaredError()

    def call(self, x, y):
        # Use the VGG feature extractor to compute features
        x_features = self.vgg(x)
        y_features = self.vgg(y)

        # Compute the feature difference between the generated and target images
        x_features = tf.stop_gradient(x_features)
        y_features = tf.stop_gradient(y_features)
        loss = self.loss(x_features, y_features)

        return loss
```

```python
model_path = '/content/drive/MyDrive/Biomedical AI/midtermModelsgen_model70.h5'

image_shape = (128,128,3)
gen_loss_fn = VGGLoss(image_shape)

# Load the saved model with custom_objects
model = tf.keras.models.load_model(model_path, {"call": gen_loss_fn.call})
```

```python
# Load the input image and resize it to 32x32
image_path = '/content/drive/MyDrive/Biomedical AI/DRUSEN-142234-12.jpeg'
input_image = cv2.imread(image_path)
input_image = cv2.resize(input_image, (32, 32))

# Preprocess the input image
input_image = np.array(input_image)
#input_image = input_image / 127.5 - 1.
input_image = input_image / 255.
input_image = np.expand_dims(input_image, axis=0)

# Use the generator to generate a new image
generated_image = model.predict(input_image)

# Postprocess the generated image
generated_image = (generated_image + 1.) * 127.5
generated_image = generated_image.astype(np.uint8)
```

```python
import matplotlib.pyplot as plt

# Define the labels
label_names = ['Input Image', 'Generated Image']

# Display the images with labels
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(8,4))
for i, ax in enumerate(axes):
    ax.imshow(input_image[i]) if i==0 else ax.imshow(generated_image[i-1])
    ax.set_title(label_names[i])
    ax.axis('off')
plt.show()
```

```python
import os
import tensorflow as tf
from PIL import Image

# Define paths to the generator model and image directories
input_dir = '/content/Data'
output_dir = '/content/Resized_Data'
target_size = (128, 128)

# Define the label names
label_names = os.listdir(input_dir)

# Loop over the subfolders of the input directory
for label_name in label_names:
    label_dir = os.path.join(input_dir, label_name)
    output_label_dir = os.path.join(output_dir, label_name)
    os.makedirs(output_label_dir, exist_ok=True)
    
    # Loop over the subfolders of the label directory
    sub_label_names = os.listdir(label_dir)
    for sub_label_name in sub_label_names:
        sub_label_dir = os.path.join(label_dir, sub_label_name)
        output_sub_label_dir = os.path.join(output_label_dir, sub_label_name)
        os.makedirs(output_sub_label_dir, exist_ok=True)
        
        # Loop over the images in the sub-label directory
        image_names = os.listdir(sub_label_dir)
        for image_name in image_names:
            # Load the input image
            input_image_path = os.path.join(sub_label_dir, image_name)
            img = tf.keras.preprocessing.image.load_img(input_image_path, target_size=(128,128))
            high_res = tf.image.convert_image_dtype(img, tf.float32)
            high_res_input = np.expand_dims(high_res, axis=0)
            
            # Use the model to generate a new image
            generated_image = model.predict(high_res_input)
            
            # Convert the generated image back to an image file
            generated_image = (generated_image + 1.) * 127.5
            generated_image = generated_image.astype(np.uint8)
            output_image_path = os.path.join(output_sub_label_dir, image_name)
            output_img = tf.keras.preprocessing.image.array_to_img(generated_image[0])
            output_img.save(output_image_path)
```
